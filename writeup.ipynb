MNIST Validation Accuracy vs. Size of Training Set

```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
xData = [i for i in range(0,15000,500)]
def func(x, e, a):
    return math.pow(x, a) * math.pow(e, 1-a)
yData = [func(i, 100, 0.3) for i in xData]
plt.plot(xData, yData, 'b-')

yData = [func(i, 100, 0.4) for i in xData]
plt.plot(xData, yData, 'g-')

yData = [func(i, 400, 0.3) for i in xData]
plt.plot(xData, yData, 'r-')

plt.title("Cobb-Douglas")
plt.axis([0, 10000, 0, 1100])
plt.xlabel('K/L, capital per labor ratio')
plt.ylabel('Y/L, output per worker ratio')
plt.show()
```

Spam Validation Accuracy vs. Training Set Size

```python
xData = [i for i in range(0,15000,500)]
def func(x, e, a):
    return math.pow(x, a) * math.pow(e, 1-a)
yData = [func(i, 1000, 0.333) for i in xData]
plt.plot(xData, yData, 'b-')

plt.title("Cobb-Douglas, E = 1000, alpha = 1/3")
plt.axis([0, 10000, 0, 2500])
plt.xlabel('K/L, capital per labor ratio')
plt.ylabel('Y/L, output per worker ratio')
plt.show()
```
